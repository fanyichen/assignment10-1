```python
import pandas as pd
import numpy as np
```

```python
# Load data
data = pd.read_csv('DOHMH_New_York_City_Restaurant_Inspection_Results.csv',sep=',',header=0)
```

```python
# Clean data
data = data.dropna(subset = ['GRADE'])
data = data[~data['GRADE'].str.contains('Not Yet Graded')]
data = data[~data['GRADE'].str.contains('P')]
data = data[~data['GRADE'].str.contains('Z')]
data_si = data['BORO']=='STATEN ISLAND'
data_bx = data['BORO']=='BRONX'
data_bk = data['BORO']=='BROOKLYN'
data_qu = data['BORO']=='QUEENS'
data_ny = data['BORO']=='MANHATTAN'
```

```python
data_si
```

```python
# Improving? 
restaurants = data['CAMIS'].unique()

def test_grades(restaurant):
    grades = ['A','B','C']
    rest_data = data[data['CAMIS'] == restaurant]
    rest_data['GRADE DATE'] = pd.to_datetime(rest_data['GRADE DATE'])
    rest_data = rest_data.sort(['GRADE DATE'])
    rest_data = rest_data.reset_index()
    first_grade = rest_data.iloc[0]['GRADE']
    last_grade = rest_data.iloc[-1]['GRADE']
    grade1 = grades.index(first_grade)
    grade2 = grades.index(last_grade)
    if grade1 == grade2:
        improve = 0
    elif grade1 < grade2:
        improve = 1
    elif grade1 > grade2:
        improve = -1
    return improve
```

```python
improve_list = []

for r in restaurants:
    improve_list.append(test_grades(r))
    
print improve_list
```
